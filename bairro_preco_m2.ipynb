# Contexto

Foi realizado o web scrapping nas principais imobiliárias de Teresina.
Com as informações de Preço e os Bairros, foi calculada a média de preço por metro2 por Bairro.

Como o valor médio do metro2 por bairro pouco se altera, normalmente, durante o ano , a coleta e atualização dos preços
pode ser realizada com essa regularidade, sempre observando as imprevibilidades do mercado. 

```python
import pandas as pd 
```

```python
df = pd.read_csv('df.csv')
```

```python
df
```

```python
df.isna().sum()
```

```python
a = df.groupby(['Bairro']).agg({'Preço': 'mean', 'Area construida': 'mean'}).reset_index()
a
#df1[df1.Bairro == 'Acarape'][]
a['preco_area'] = a.Preço/a['Area construida']
```

```python
a[['Bairro', 'preco_area']].to_csv('Bairro_valor_metro2.csv')
```
